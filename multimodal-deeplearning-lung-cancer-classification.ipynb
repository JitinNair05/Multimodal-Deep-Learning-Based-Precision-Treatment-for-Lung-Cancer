```python
# ============================================================
# FINAL IEEE MULTIMODAL LUNG CANCER PIPELINE
# HIGH ACCURACY VERSION (SAME METHODOLOGY)
# NOW WITH PRECISION, RECALL, CONFUSION MATRICES, t-SNE & WEIGHT VISUALIZATION
# ============================================================

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.manifold import TSNE  # <-- ADDED FOR t-SNE

from sklearn.metrics import (
    accuracy_score,
    f1_score,
    precision_score,
    recall_score,
    confusion_matrix,
    roc_curve
)

print("TensorFlow:", tf.__version__)

# ============================================================
# PATH (UNCHANGED)
# ============================================================

BASE_PATH = "/kaggle/input/datasets/jitinrnair/newdataset/roi_ready-20260216T144003Z-1-001"
CSV_PATH = "/kaggle/input/datasets/jitinrnair/newdataset/final_multimodal_table.csv"

IMG = 224
EPOCH = 30
BATCH = 16

CLIN = ["Age","weight (kg)","Sex","Smoking History","T-Stage","N-Stage","M-Stage"]

CAT = ["Sex","Smoking History","T-Stage","N-Stage","M-Stage"]
NUM = ["Age","weight (kg)"]

# ============================================================
# CALLBACKS
# ============================================================

callbacks=[
    keras.callbacks.EarlyStopping(
        monitor="val_accuracy",
        patience=5,
        restore_best_weights=True
    ),
    keras.callbacks.ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.3,
        patience=3
    )
]

# ============================================================
# LOAD DATA
# ============================================================

df = pd.read_csv(CSV_PATH)

df["image_path"]=df["image_path"].str.replace("\\","/",regex=False)
df["image_path"]=df["image_path"].str.replace("../","",regex=False)

df["image_path"]=BASE_PATH+"/"+df["image_path"]

# ============================================================
# ENCODE CLINICAL
# ============================================================

for c in CAT:
    df[c]=LabelEncoder().fit_transform(df[c])

df[NUM]=StandardScaler().fit_transform(df[NUM])

lab=LabelEncoder()

y=lab.fit_transform(df["label"])

NCLASS=len(lab.classes_)

Xclin=df[CLIN].values

paths=df["image_path"].values

# ============================================================
# LOAD CT
# ============================================================

def load_images(paths):
    images=[]
    for p in paths:
        img=tf.io.read_file(p)
        img=tf.image.decode_png(img,channels=3)
        img=tf.image.resize(img,(IMG,IMG))
        img=keras.applications.efficientnet.preprocess_input(img)
        images.append(img.numpy())
    return np.array(images)

Xct=load_images(paths)

# ============================================================
# SPLIT
# ============================================================

Xct_train,Xct_test,Xclin_train,Xclin_test,y_train,y_test=train_test_split(
    Xct,Xclin,y,test_size=0.2,random_state=42,stratify=y
)

# ============================================================
# CT MODELS
# ============================================================

CT_MODELS={
    "VGG16":keras.applications.VGG16,
    "ResNet50":keras.applications.ResNet50,
    "DenseNet121":keras.applications.DenseNet121,
    "EfficientNetB0":keras.applications.EfficientNetB0
}

ct_results={}
ct_features={}
ct_predictions={} # <-- ADDED to store predictions for Confusion Matrices

# ============================================================
# TRAIN CT
# ============================================================

for name,model_class in CT_MODELS.items():
    print("\nTraining CT:",name)
    base=model_class(
        include_top=False,
        weights="imagenet",
        input_shape=(IMG,IMG,3)
    )

    base.trainable=False

    x=layers.GlobalAveragePooling2D()(base.output)
    x=layers.BatchNormalization()(x)
    x=layers.Dense(256,activation="relu")(x)
    x=layers.Dropout(0.5)(x)

    feat=x

    out=layers.Dense(NCLASS,activation="softmax")(x)

    model=keras.Model(base.input,out)
    feature_model=keras.Model(base.input,feat)

    model.compile(
        optimizer=keras.optimizers.Adam(0.0001),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )

    model.fit(
        Xct_train,
        y_train,
        epochs=EPOCH,
        batch_size=BATCH,
        validation_split=0.2,
        callbacks=callbacks
    )

    pred=model.predict(Xct_test)
    pred_label=np.argmax(pred,1)
    
    ct_predictions[name] = pred_label # <-- STORED for CM

    acc=accuracy_score(y_test,pred_label)
    f1=f1_score(y_test,pred_label,average="macro")
    precision=precision_score(y_test,pred_label,average="macro")
    recall=recall_score(y_test,pred_label,average="macro")

    ct_results[name]=[acc,f1,precision,recall]
    ct_features[name]=feature_model.predict(Xct_test)

# ============================================================
# BEST CT
# ============================================================

ct_results=pd.DataFrame(
    ct_results,
    index=["Accuracy","F1","Precision","Recall"]
).T

print("\nCT RESULTS")
print(ct_results)

best_ct=ct_results["Accuracy"].idxmax()
best_ct_features=ct_features[best_ct]

# ============================================================
# CLINICAL MODEL
# ============================================================

print("\nTraining Clinical")

clin_in=keras.Input((len(CLIN),))
x=layers.Dense(128,activation="relu")(clin_in)
x=layers.BatchNormalization()(x)
x=layers.Dropout(0.3)(x)

feat=x

out=layers.Dense(NCLASS,activation="softmax")(x)

clin_model=keras.Model(clin_in,out)
clin_feat_model=keras.Model(clin_in,feat) # <-- ADDED for t-SNE

clin_model.compile(
    optimizer=keras.optimizers.Adam(0.0001),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

clin_model.fit(
    Xclin_train,
    y_train,
    epochs=EPOCH,
    batch_size=BATCH,
    validation_split=0.2,
    callbacks=callbacks
)

clin_pred=clin_model.predict(Xclin_test)
clin_pred_label=np.argmax(clin_pred,1)

clin_features_test = clin_feat_model.predict(Xclin_test) # <-- EXTRACTED for t-SNE

clin_acc=accuracy_score(y_test,clin_pred_label)
clin_f1=f1_score(y_test,clin_pred_label,average="macro")
clin_precision=precision_score(y_test,clin_pred_label,average="macro")
clin_recall=recall_score(y_test,clin_pred_label,average="macro")

# ============================================================
# FUSION MODEL
# ============================================================

print("\nTraining Fusion")

img_in=keras.Input((IMG,IMG,3))
base=CT_MODELS[best_ct](include_top=False,weights="imagenet",input_tensor=img_in)
base.trainable=False
img_feat=layers.GlobalAveragePooling2D()(base.output)

clin_in=keras.Input((len(CLIN),))
clin_feat=layers.Dense(128,activation="relu")(clin_in)

fusion=layers.concatenate([img_feat,clin_feat])
fusion=layers.BatchNormalization()(fusion)
fusion=layers.Dense(256,activation="relu")(fusion)
fusion=layers.Dropout(0.5)(fusion)

out=layers.Dense(NCLASS,activation="softmax")(fusion)

fusion_model=keras.Model([img_in,clin_in],out)
fusion_feat_model=keras.Model([img_in,clin_in],fusion) # <-- ADDED for t-SNE

fusion_model.compile(
    optimizer=keras.optimizers.Adam(0.0001),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

fusion_model.fit(
    [Xct_train,Xclin_train],
    y_train,
    epochs=EPOCH,
    batch_size=BATCH,
    validation_split=0.2,
    callbacks=callbacks
)

fusion_pred=fusion_model.predict([Xct_test,Xclin_test])
fusion_pred_label=np.argmax(fusion_pred,1)

fusion_features_test = fusion_feat_model.predict([Xct_test,Xclin_test]) # <-- EXTRACTED for t-SNE

fusion_acc=accuracy_score(y_test,fusion_pred_label)
fusion_f1=f1_score(y_test,fusion_pred_label,average="macro")
fusion_precision=precision_score(y_test,fusion_pred_label,average="macro")
fusion_recall=recall_score(y_test,fusion_pred_label,average="macro")

# ============================================================
# FINAL TABLE
# ============================================================

final=pd.DataFrame({
    "Model":["Clinical",best_ct,"Fusion"],
    "Accuracy":[clin_acc,ct_results.loc[best_ct,"Accuracy"],fusion_acc],
    "Precision":[clin_precision,ct_results.loc[best_ct,"Precision"],fusion_precision],
    "Recall":[clin_recall,ct_results.loc[best_ct,"Recall"],fusion_recall],
    "F1":[clin_f1,ct_results.loc[best_ct,"F1"],fusion_f1]
})

print("\nFINAL RESULTS")
print(final)
final.to_csv("/kaggle/working/final_results.csv")

plt.figure()
final.plot(x="Model",y="Accuracy",kind="bar")
plt.title("Final Comparison")
plt.show()

# ============================================================
# NEW VISUALIZATIONS: CONFUSION MATRICES, t-SNE, AND WEIGHTS
# ============================================================

# --- 1. Confusion Matrix Helper ---
def plot_cm(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=lab.classes_, yticklabels=lab.classes_)
    plt.title(f"Confusion Matrix: {title}")
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

# --- 2. t-SNE Helper ---
def plot_tsne(features, labels, title):
    # Adjust perplexity to prevent errors on small test sets
    n_samples = features.shape[0]
    perplexity = min(30, n_samples - 1)
    
    tsne = TSNE(n_components=2, perplexity=perplexity, random_state=42)
    tsne_results = tsne.fit_transform(features)
    
    plt.figure(figsize=(8,6))
    sns.scatterplot(
        x=tsne_results[:,0], y=tsne_results[:,1],
        hue=[lab.classes_[i] for i in labels],
        palette="tab10", legend="full", alpha=0.8
    )
    plt.title(f"t-SNE Visualization: {title}")
    plt.show()

print("\n--- GENERATING CONFUSION MATRICES ---")
for name, pred_labels in ct_predictions.items():
    plot_cm(y_test, pred_labels, f"CT Backbone: {name}")

plot_cm(y_test, clin_pred_label, "Clinical Model")
plot_cm(y_test, fusion_pred_label, "Multimodal Fusion Model")

print("\n--- GENERATING t-SNE VISUALIZATIONS ---")
plot_tsne(best_ct_features, y_test, f"Best CT ({best_ct}) Features")
plot_tsne(clin_features_test, y_test, "Clinical Features")
plot_tsne(fusion_features_test, y_test, "Multimodal Fusion Features")

print("\n--- VISUALIZING CLINICAL FEATURE WEIGHTS ---")
# Extracting the weights from the first dense layer of the clinical model
# clin_model summary: InputLayer(0), Dense(128)(1), BatchNormalization(2), Dropout(3), Dense(NCLASS)(4)
clinical_dense_layer = clin_model.layers[1]
weights, biases = clinical_dense_layer.get_weights()

plt.figure(figsize=(12, 6))
# Using absolute weights to highlight feature importance magnitude
sns.heatmap(np.abs(weights), cmap='viridis', yticklabels=CLIN, 
            cbar_kws={'label': 'Absolute Weight Magnitude'})
plt.title('Learned Clinical Feature Representation Weights (128 Hidden Units)')
plt.xlabel('Hidden Units')
plt.ylabel('Clinical Features')
plt.tight_layout()
plt.show()
```

```python
# ============================================================
# ROC CURVE COMPARISON (EFFICIENTNET, CLINICAL, FUSION)
# ============================================================
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
import numpy as np
import matplotlib.pyplot as plt

# 1. Binarize the true labels for multiclass ROC calculation
y_test_bin = label_binarize(y_test, classes=range(NCLASS))

# 2. Get probability predictions (not argmax labels)
# Note: 'model' holds EfficientNetB0 as it was the last CT model in the loop
eff_prob = model.predict(Xct_test) 
clin_prob = clin_model.predict(Xclin_test)
fusion_prob = fusion_model.predict([Xct_test, Xclin_test])

# 3. Helper function to calculate Macro-Average ROC
def get_macro_roc(y_true, y_prob):
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    
    # Calculate ROC for each class
    for i in range(NCLASS):
        fpr[i], tpr[i], _ = roc_curve(y_true[:, i], y_prob[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
        
    # Aggregate all false positive rates
    all_fpr = np.unique(np.concatenate([fpr[i] for i in range(NCLASS)]))
    
    # Interpolate all ROC curves at these points
    mean_tpr = np.zeros_like(all_fpr)
    for i in range(NCLASS):
        mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])
        
    # Average it and compute AUC
    mean_tpr /= NCLASS
    macro_fpr = all_fpr
    macro_tpr = mean_tpr
    macro_auc = auc(macro_fpr, macro_tpr)
    
    return macro_fpr, macro_tpr, macro_auc

# Calculate Macro-ROC for the three targeted models
fpr_eff, tpr_eff, auc_eff = get_macro_roc(y_test_bin, eff_prob)
fpr_clin, tpr_clin, auc_clin = get_macro_roc(y_test_bin, clin_prob)
fpr_fusion, tpr_fusion, auc_fusion = get_macro_roc(y_test_bin, fusion_prob)

# 4. Plot the comparison
plt.figure(figsize=(9, 7))
plt.plot(fpr_eff, tpr_eff, label=f'EfficientNetB0 (AUC = {auc_eff:.3f})', color='royalblue', lw=2)
plt.plot(fpr_clin, tpr_clin, label=f'Clinical Model (AUC = {auc_clin:.3f})', color='forestgreen', lw=2)
plt.plot(fpr_fusion, tpr_fusion, label=f'Fusion Model (AUC = {auc_fusion:.3f})', color='crimson', lw=2)

# Diagonal chance line
plt.plot([0, 1], [0, 1], 'k--', lw=2, alpha=0.6) 

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (FPR)', fontsize=12)
plt.ylabel('True Positive Rate (TPR)', fontsize=12)
plt.title('Multiclass Macro-Average ROC Curve Comparison', fontsize=14, fontweight='bold')
plt.legend(loc="lower right", fontsize=11)
plt.grid(alpha=0.3)
plt.show()
```

```python
# ============================================================
# INDIVIDUAL CLASS ROC CURVES (EFFICIENTNET, CLINICAL, FUSION)
# ============================================================
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Note: This relies on y_test_bin, eff_prob, clin_prob, and fusion_prob 
# generated in the previous Macro-ROC cell.

def plot_per_class_roc(y_true_bin, y_prob, model_name, ax):
    # Standard color palette for distinct classes
    colors = ['blue', 'green', 'red', 'purple', 'orange', 'brown', 'pink', 'gray']
    
    for i in range(NCLASS):
        fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_prob[:, i])
        roc_auc = auc(fpr, tpr)
        
        class_name = lab.classes_[i]
        color = colors[i % len(colors)]
        
        ax.plot(fpr, tpr, color=color, lw=2,
                label=f'{class_name} (AUC = {roc_auc:.2f})')
        
    ax.plot([0, 1], [0, 1], 'k--', lw=2, alpha=0.5)
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate', fontsize=11)
    ax.set_ylabel('True Positive Rate', fontsize=11)
    ax.set_title(f'{model_name} ROC', fontsize=13, fontweight='bold')
    ax.legend(loc="lower right", fontsize=9)
    ax.grid(alpha=0.3)

# Create a 1x3 figure to display all three models side-by-side
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

plot_per_class_roc(y_test_bin, eff_prob, "EfficientNetB0", axes[0])
plot_per_class_roc(y_test_bin, clin_prob, "Clinical Model", axes[1])
plot_per_class_roc(y_test_bin, fusion_prob, "Fusion Model", axes[2])

# Add a main title for the entire figure
plt.suptitle('Per-Class ROC Curves Across Modalities', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```
